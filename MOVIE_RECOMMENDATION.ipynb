```python
# Import pandas library for data manipulation and analysis
import pandas as pd

# pd.read_csv() loads the CSV file into a pandas DataFrame
ratings = pd.read_csv(
    r"C:\Users\B.SARAN PYARI\Downloads\ml-latest-small\ml-latest-small\ratings.csv"
)

# Display the first 5 rows of the dataset using head()
print(ratings.head())
```

```python
movies = pd.read_csv(
    r"C:\Users\B.SARAN PYARI\Downloads\ml-latest-small\ml-latest-small\movies.csv"
)

print(movies.head())
```

```python
# The merge is done using the common column "movieId" .This helps us combine user ratings with movie information (title, genres)

data = pd.merge(ratings, movies, on="movieId")

# Display the first 5 rows of the merged dataset using head()
print(data.head())
```

```python
# Create a pivot table (User–Movie Matrix)

movie_matrix = data.pivot_table(
    index="userId",
    columns="title",
    values="rating"
)

# Display the first 5 rows of the user–movie matrix
print(movie_matrix.head())
```

```python
# groupby('title') groups the data by movie title and count() counts the number of ratings for each movie
movie_ratings_count = data.groupby('title')['rating'].count()

# mean() computes the average rating given by all users for each movie
ratings_mean = data.groupby('title')['rating'].mean()

# Create a new DataFrame using the average ratings
ratings_df = pd.DataFrame(ratings_mean)

# Add a new column 'num_ratings' to store how many ratings each movie has . This helps identify popular movies
ratings_df['num_ratings'] = movie_ratings_count

# Display the first 5 rows of the ratings DataFrame
print(ratings_df.head())
```

```python
import pandas as pd  # import pandas for data handling

# load ratings dataset
ratings = pd.read_csv(r"C:\Users\B.SARAN PYARI\Downloads\ml-latest-small\ml-latest-small\ratings.csv")  
# load movies dataset
movies = pd.read_csv(r"C:\Users\B.SARAN PYARI\Downloads\ml-latest-small\ml-latest-small\movies.csv")
# remove year from movie titles
movies["title"] = movies["title"].str.replace(r"\(\d{4}\)", "", regex=True).str.strip()  
# merge ratings and movies using movieId
data = pd.merge(ratings, movies, on="movieId")

print(data.head())
```

```python
def recommend_movies(movie_name):
    # find movies matching the entered name
    matches = [title for title in movie_matrix.columns if movie_name.lower() in title.lower()]  
    # check if no movie matches the input
    if len(matches) == 0:
        print("Movie not found")
        return
    # take the first matched movie title
    movie_title = matches[0]

    # get ratings of the selected movie from the matrix
    movie_ratings = movie_matrix[movie_title]
    # compute correlation between this movie and all other movies
    similar_movies = movie_matrix.corrwith(movie_ratings)
    # convert correlation results into a DataFrame
    corr_movies = pd.DataFrame(similar_movies, columns=['correlation'])
    # remove movies with no correlation values
    corr_movies.dropna(inplace=True)
    # sort movies by similarity and select top 10 recommendations
    recommendations = corr_movies.sort_values(
        'correlation', ascending=False
    )[1:11]  
    # display the matched movie
    print("\nMovie found:", movie_title)
    # heading for recommendations
    print("\nRecommended Movies:\n")
    # print recommended movies
    print(recommendations)  
```

```python
movie = input("Enter movie name: ")
recommend_movies(movie)
```
